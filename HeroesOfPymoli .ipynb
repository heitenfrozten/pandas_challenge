```python
#Bins creation

bins = [0, 11, 15, 19,23,27,31,35,39,43,47,51,55,59]

# Create the names for the four bins
group_names = ["a", "b", "c", "d", "e","f","g","h","i","j","k","l","m"]

purchase_df["clasificacion"] = pd.cut(purchase_df["Age"], bins, labels=group_names)
purchase_df
```

```python
count_class = purchase_df["clasificacion"].value_counts()
count_class
```

```python
purchase_df_2 = purchase_df.groupby("clasificacion")
purchase_df_2.head()
```

```python
# Find how many rows fall into each bin
print(purchase_df_2["clasificacion"].count())

# Get the average of each column within the GroupBy object
purchase_df_2[["Price"]].mean()

#statistics by bin
```
